```python
# Reverse a number using while loop
n = int(input("Enter a number: ")) 
rev = 0
temp = n  

while n > 0:
    digit = n % 10    
    rev = rev * 10 + digit 
    n = n // 10       

print(f"Reverse of {temp} = {rev}")
```

```python
import math

# Check if a number is prime using while loop
n = int(input("Enter a number: "))

if n <= 1:
    print(f"{n} is not prime")
else:
    i = 2
    is_prime = True
    
    while i <= math.sqrt(n):  
        if n % i == 0:       
            is_prime = False
            break
        i = i + 1
    
    if is_prime:
        print(f"{n} is prime")
    else:
         print(f"{n} is not prime")


    
   
```
