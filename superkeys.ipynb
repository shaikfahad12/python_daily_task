```python
from abc import ABC, abstractmethod
import math

# 1. Abstract Class
class Shape(ABC):
    
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


# 2. Subclass: Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


# 3. Subclass: Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


# Testing
c1 = Circle(5)
r1 = Rectangle(4, 6)

print(f"Circle - Radius: 5")
print(f"Area: {c1.area():.2f}")
print(f"Perimeter: {c1.perimeter():.2f}")
print("-" * 20)

print(f"Rectangle - Length: 4, Width: 6")
print(f"Area: {r1.area()}")
print(f"Perimeter: {r1.perimeter()}")
```

```python
#2. Class Name: Vehicle
#Problem
#reate an abstract class named Vehicle.
#Explanation
#Every vehicle can start and stop, but each vehicle does it differently.
#Create two abstract methods:
#• start()
#• stop()
#Create a Car class that implements both methods.

from abc import ABC, abstractmethod

# 1. Abstract Class
class Vehicle(ABC):
    
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass


# 2. Subclass: Car
class Car(Vehicle):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        self.is_running = False

    def start(self):
        if not self.is_running:
            self.is_running = True
            print(f"{self.brand} {self.model} engine started. Vroom!")
        else:
            print(f"{self.brand} {self.model} is already running.")

    def stop(self):
        if self.is_running:
            self.is_running = False
            print(f"{self.brand} {self.model} engine stopped.")
        else:
            print(f"{self.brand} {self.model} is already stopped.")


# Testing
car1 = Car("Tesla", "Model 3")
car2 = Car("Maruti", "Swift")

car1.start()   # Tesla Model 3 engine started. Vroom!
car1.stop()    # Tesla Model 3 engine stopped.

car2.start()   # Maruti Swift engine started. Vroom!
car2.start()   # Maruti Swift is already running.
```

```python
from abc import ABC, abstractmethod

# 1. Abstract Class
class Animal(ABC):
    
    @abstractmethod
    def make_sound(self):
        pass


# 2. Subclass: Dog
class Dog(Animal):
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print(f"{self.name} says: Woof Woof!")


# 3. Subclass: Cat
class Cat(Animal):
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print(f"{self.name} says: Meow Meow!")


# Testing
dog1 = Dog("Tommy")
cat1 = Cat("Kitty")

dog1.make_sound()  # Tommy says: Woof Woof!
cat1.make_sound()  # Kitty says: Meow Meow!


# Bonus: Polymorphism with a list
animals = [Dog("Sheru"), Cat("Luna"), Dog("Bruno")]

print("\nAll animals making sound:")
for animal in animals:
    animal.make_sound()
```

```python
from abc import ABC, abstractmethod

# 1. Abstract Class
class Payment(ABC):
    
    @abstractmethod
    def process_payment(self, amount):
        pass


# 2. Subclass: CreditCard
class CreditCard(Payment):
    def __init__(self, card_number, card_holder):
        self.card_number = card_number
        self.card_holder = card_holder

    def process_payment(self, amount):
        if amount > 0:
            print(f"Processing Credit Card payment of Rs.{amount}")
            print(f"Card: **** {self.card_number[-4:]}")
            print(f"Card Holder: {self.card_holder}")
            print("Payment Successful via Credit Card!")
        else:
            print("Invalid amount")


# 3. Subclass: Cash
class Cash(Payment):
    def __init__(self, received_by):
        self.received_by = received_by

    def process_payment(self, amount):
        if amount > 0:
            print(f"Processing Cash payment of Rs.{amount}")
            print(f"Cash received by: {self.received_by}")
            print("Payment Successful via Cash!")
        else:
            print("Invalid amount")


# Testing
p1 = CreditCard("1234567890123456", "Fahad")
p2 = Cash("Shop Owner")

print("--- Credit Card Payment ---")
p1.process_payment(1500)

print("\n--- Cash Payment ---")
p2.process_payment(500)


# Bonus: Polymorphism
print("\n--- Processing Multiple Payments ---")
payments = [
    CreditCard("9876543210987654", "fahad"),
    Cash("Counter 1")
]

for payment in payments:
    payment.process_payment(1000)
```

```python
from abc import ABC, abstractmethod
from datetime import datetime

# 1. Abstract Class
class Report(ABC):
    
    @abstractmethod
    def generate(self, data):
        pass


# 2. Subclass: PDFReport
class PDFReport(Report):
    def generate(self, data):
        print("Generating PDF Report...")
        print(f"Title: {data['title']}")
        print(f"Date: {datetime.now().strftime('%Y-%m-%d')}")
        print("Content:", data['content'])
        print("PDF Report saved successfully!")


# 3. Subclass: ExcelReport
class ExcelReport(Report):
    def generate(self, data):
        print("Generating Excel Report...")
        print(f"Sheet Name: {data['title']}")
        print("Rows:", len(data['content']))
        print("Columns: Name, Value")
        print("Excel Report saved successfully! ")


# Testing
report_data = {
    'title': 'Monthly Sales',
    'content': ['Product A: 100', 'Product B: 200', 'Product C: 150']
}

pdf = PDFReport()
excel = ExcelReport()

print("--- PDF Report ---")
pdf.generate(report_data)

print("\n--- Excel Report ---")
excel.generate(report_data)


# Bonus: Polymorphism
print("\n--- Generating all reports ---")
reports = [PDFReport(), ExcelReport()]

for report in reports:
    report.generate(report_data)
    print("-" * 30)
```
