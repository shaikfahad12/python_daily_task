```python
#Define class Car with attributes brand, model, and year. Create an object and display all attributes.

class car:
    def __init__(self,brands,model,year):
        self.brands = brands
        self.model = model
        self.year = year

        # display
    def display_car(self):
        print(f"car brand:{self.brands}")
        print(f"car model:{self.model}")
        print(f"car year:{self.year}")

        #object 
car = car("landcrusier","toyota",2026)
car.display_car()
```

```python
#3. Create class Book with attributes title, author, price. Add a method display_info().
class book:
        def __init__(self,title,author,price):
            self.title = title
            self.author = author
            self.price = price

        # display
        def display_info(self):
            print(f"info title:{self.title}")
            print(f"info author:{self.author}")
            print(f"info price:{self.price}")

        #object 
book = book("cosm","fahad",450)
book.display_info()
```

```python
#Create class Employee with attributes emp_id, name, salary. Create 3 objects.
class employee:
    def __init__(self,emp_id,name,salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

     #display
    def display_employee(self): 
         print(f"employee:{self.emp_id}")
         print(f"employee:{self.name}")
         print(f"employee:{self.salary}")


         #object
emp = employee(1234,"fahad",1500000)
emp.display_employee()
```

```python
#Create class Rectangle with attributes length and width. Add method to calculate area

class Rectangle:
    def __init__(self,length,width):
        self.length = length
        self.width = width

        #display
    def calculate_area(self):
        area=self.length * self.width
        print(area)

#object 
rec=Rectangle(2,4)
rec.calculate_area()

        
    
```

```python
#7. Define class Person with attributes name and city. Create object and change city value
class person:
    def __init__(self,name,city):
        self.name = name
        self.city = city

    def update_city(self,new_city):
        self.city = new_city
       

    def display(self):
        print(f"name:{self.city}")
        print(f"new_city{self.city}")

p = person("fahad","hyderabad")
p.display()
print()
p.update_city("mumbai")
p.display()
```
